1)Ekteki veri dosyasını okuyup, veriyi adı adult_df olan DataFrame’de tutunuz.

```python
import pandas as pd
import numpy as np
adult_df=pd.read_csv("adult.csv",sep=",")
adult_df
```

2) '?' şeklinde var olan kayıtları np.nan (Numpy NaN) olarak değiştiriniz.

```python
adult_df.replace('?',np.nan,inplace=True)
adult_df
```

3) Eksik veri içeren sütunları siliniz.

```python
adult_df=adult_df.dropna(axis=1)
adult_df
```

4) Kadınların yüzde kaçının yıllık kazancı 50000 dolardan fazladır?

```python
len_w=len(adult_df.loc[(adult_df['gender']=='Female')])   #14695 kişi kadın
len_w50=len(adult_df.loc[(adult_df['gender']=='Female')&(adult_df['income']=='>50K')])  #1669 kişi kadın ve >50K dolar kazanıyor
percent_w=(len_w50*100)/len_w
print("Kadınların %{:.2f}\'nin yıllık kazancı 50000 dolardan fazladır.".format(percent_w))
```
